```python
df = pd.read_csv('PhoneReviews.csv')
```

```python
df.drop(columns=['Product Name', 'Brand Name', 'Price', 'Rating', 'Review Votes'], axis=1, inplace=True)
```

Use lambda function to apply Vader sentiment to the text ot calculate Sentiment Intensity of all reviews, add the new column to the dataframe as scores

```python
analyzer = SentimentIntensityAnalyzer()

df['neg'] = df['Reviews'].apply(lambda x:analyzer.polarity_scores(x)['neg'])
df['neu'] = df['Reviews'].apply(lambda x:analyzer.polarity_scores(x)['neu'])
df['pos'] = df['Reviews'].apply(lambda x:analyzer.polarity_scores(x)['pos'])
df['scores'] = df['Reviews'].apply(lambda x:analyzer.polarity_scores(x)['compound'])
```

```python
df
```

Add sentiment type by checking the coupund values (>0 Positive, <0 Negative, ==0 Neutral)

```python
df['compound'] = df['Reviews'].apply(lambda x:analyzer.polarity_scores(x)['compound'])
```

```python
df['sentiment_type'] = df['scores'].apply(lambda c: 'Positive' if c>0 else ('Negative' if c<0 else 'Neutral'))
df.drop(['compound'], axis=1, inplace=True)
df
```
